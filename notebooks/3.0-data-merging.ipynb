## Xử lý df2 và df3

```python
df2 = df2[['prompt','label_unsafe']]
df3 = df3[['prompt','label_unsafe']]
```

## Gộp dữ liệu

```python
master_df = pd.concat([df1,df2,df3,df4], ignore_index=True)
```

```python
duplicates = master_df[master_df.duplicated(subset=["prompt"], keep=False)]
```

```python
master_df = master_df.drop_duplicates()
```

```python
master_df.duplicated().sum()
```

```python
master_df.shape
```

```python
master_df["label_unsafe"].value_counts(normalize=True) * 100
```

```python
safe = master_df[master_df["label_unsafe"] == 0]
unsafe = master_df[master_df["label_unsafe"] == 1]
```

```python
min_size = min(len(safe), len(unsafe))
```

```python
safe = safe.sample(n=min_size, random_state=42)
unsafe = unsafe.sample(n=min_size, random_state=42)
```

```python
balanced_df = pd.concat([safe, unsafe], ignore_index=True)
```

```python
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
balanced_df["label_unsafe"].value_counts(normalize=True) * 100
```

## Shuffle dữ liệu

```python
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

## Chia 80% → train_master.csv và 20% → val_master.csv

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, val_df = train_test_split(
    balanced_df,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=balanced_df["label_unsafe"]
)
```

```python
len(train_df)
```

```python
len(val_df)
```

```python
train_df["label_unsafe"].value_counts(normalize=True) * 100
```

```python
val_df["label_unsafe"].value_counts(normalize=True) * 100
```

```python
train_df.to_csv(
    "../data/processed/train_master.csv",
    index=False
)
```

```python
val_df.to_csv(
    "../data/processed/val_master.csv",
    index=False
)
```
