```python

df_balanced["label_unsafe"] = (
    (df_balanced["toxicity"] == 1) | (df_balanced["jailbreaking"] == 1)
).astype(int)

df_balanced["label_unsafe"].value_counts()
df_balanced['label_unsafe'].mean()*100
(1 - df_balanced['label_unsafe'].mean()) * 100


pd.crosstab(
    df_balanced["toxicity"],
    df_balanced["jailbreaking"],
    rownames=["toxicity"],
    colnames=["jailbreaking"],
    margins=True
)
```

```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r"https?://\S+|www\.\S+", "", text)
    text = re.sub(r"[\x00-\x08\x0b\x0c\x0e-\x1f\x7f]", "", text)
    text = re.sub(r"[\r\n\t]+", " ", text)
    text = re.sub(r" {2,}", " ", text).strip()
    return text


def is_garbage(text, min_words=3, min_chars=10):
    if not isinstance(text, str) or len(text.strip()) < min_chars:
        return True
    if len(text.strip().split()) < min_words:
        return True
    if not re.search(r"[a-zA-Z\u00C0-\u024F\u0400-\u04FF]", text):
        return True
    return False


```

```python
n_before = len(df_balanced)

df_balanced["prompt"]       = df_balanced["prompt"].apply(clean_text)
df_balanced["model_output"] = df_balanced["model_output"].apply(clean_text)

mask_garbage = df_balanced["prompt"].apply(is_garbage)
n_garbage    = mask_garbage.sum()
df_balanced  = df_balanced[~mask_garbage].reset_index(drop=True)


df_balanced["label_unsafe"].value_counts()
```

```python
cols = ["prompt", "model_output", "toxicity", "jailbreaking", "label_unsafe"]

df_balanced[df_balanced["label_unsafe"] == 0][cols].head(3)

df_balanced[df_balanced["label_unsafe"] == 1][cols].head(3)
```

```python
import os

output_path = "../data/processed/toxic_chat_processed.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

cols_out = ["conv_id", "prompt", "model_output", "label_unsafe", "toxicity", "jailbreaking"]
df_out   = df_balanced[[c for c in cols_out if c in df_balanced.columns]]

df_out.to_csv(output_path, index=False, encoding="utf-8")


df_out["label_unsafe"].value_counts()
```
